```python
#1.Write a Pandas program to compare the elements of two Pandas series. Sample Series [4,65,436,3,9], [7,0,3,897,9]
#using comparison operator 
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])

comparison = series1 == series2

print(comparison)
```

```python
#2.	Write a Pandas program to add, subtract, multiply and divide two Pandas series. Sample Series: [2,4,6,8,14], [1,3,5,7,9]
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

# Addition
addition = series1 + series2
print("Addition:")
print(addition)

# Subtraction
subtraction = series1 - series2
print("\nSubtraction:")
print(subtraction)

# Multiplication
multiplication = series1 * series2
print("\nMultiplication:")
print(multiplication)

# Division
division = series1 / series2
print("\nDivision:")
print(division)
```

```python
#3.	Write a Pandas program to convert a dictionary to a Pandas series. Sample dictionary: dictionary1 = {‘Josh’: 24, ‘Sam’: 36, ‘Peace’: 19, ‘Charles’: 65, ‘Tom’: 44}
dictionary1 = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}
series = pd.Series(dictionary1)
print(series)
```

```python
#4.	Write a Pandas program to convert a given series to an array.Sample series: [‘Love’, 800, ‘Joy’, 789.9, ‘Peace’, True]
series = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])

array = np.array(series)

print(array)
```

```python

df = pd.read_csv('AfricaCupofNationsMatches.csv')

# Creating a Series from the HomeTeamGoals column
series = df['HomeTeamGoals']

# Finding the most frequent value
most_frequent = series.value_counts().idxmax()

print("Most frequent value:", most_frequent)

# Replacing other values with 'Other'
series[series != most_frequent] = 'Other'

# Displaying the modified Series
print(series)
```
